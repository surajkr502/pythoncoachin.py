```python
import matplotlib.pyplot as plt
import math

# Input sides of the triangle
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

# Calculate semi-perimeter
s = (a + b + c) / 2

# Calculate area using Heron's formula
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area of the triangle is:", area)

# For plotting, we need coordinates
# We'll place first vertex at (0,0), second at (c,0)
# Third vertex (x3, y3) calculated using distance formula
x1, y1 = 0, 0
x2, y2 = c, 0

# Using formula for coordinates of third vertex
# a^2 = (x3 - x2)^2 + (y3 - y2)^2
# b^2 = (x3 - x1)^2 + (y3 - y1)^2
x3 = (b**2 + c**2 - a**2) / (2 * c)
y3 = math.sqrt(max(b**2 - x3**2, 0))  # ensure non-negative

# Triangle vertices
triangle_x = [x1, x2, x3, x1]
triangle_y = [y1, y2, y3, y1]

# Plot the triangle
plt.figure(figsize=(6,6))
plt.plot(triangle_x, triangle_y, 'b-', linewidth=2)  # triangle edges
plt.fill(triangle_x, triangle_y, color='skyblue', alpha=0.5)  # fill color
plt.scatter([x1, x2, x3], [y1, y2, y3], color='red')  # vertices

# Label vertices
plt.text(x1, y1, 'A', fontsize=12, color='red', ha='right')
plt.text(x2, y2, 'B', fontsize=12, color='red', ha='left')
plt.text(x3, y3, 'C', fontsize=12, color='red', ha='center')

plt.title(f"Triangle with Area = {area:.2f}")
plt.axis('equal')  # keep aspect ratio
plt.grid(True)
plt.show()
```
