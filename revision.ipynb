### 24. What is the list in python? Explain list methods with examples. 
 We use a list in Python when we want to store multiple objects together in an ordered and mutable collection that also allows duplicate values.
* Whenever we want to create a group of objects where we want below mention
properties, then we are using list sequence.
1. Duplicates are allowed.
2. Order is preserved.
3. Objects are mutable.
4. Indexing are allowed.
5. Slicing are allowed.
6. Represented in square bracket with comma separated objects.
7. Homogeneous and Heterogeneous both objects are allowed.

```python
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

print(factorial(5))
```

```python
def prime(n):
    if n <= 1:
        return "Not Prime"

    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"

    return "Prime"

print(prime(7))
```

## 26. What is the dictionary in python? Explain methods with examples. 
* I we want to represent the group of object in key value pair then we use dictionary.
### Characteristics of Dictionary
1. Dictionary will contain data in the form of key, value pairs.
2. Key and values are separated by a colon “:” symbol
3. One key-value pair can be represented as an item.
4. Duplicate keys are not allowed.
5. Duplicate values can be allowed.
6. Heterogeneous objects are allowed for both keys and values.
7. Insertion order is not preserved.
8. Dictionary object having mutable nature.
9. Dictionary objects are dynamic.
10. Indexing and slicing concepts are not applicable

```python
def palindrome(text):

    if text == text[::-1]:
        return "Palindrome"

    return "Not Palindrome"

print(palindrome("madam"))
```

```python
def upper_count(text):

    count = 0

    for i in text:
        if i.isupper():
            count += 1

    return count

print(upper_count("PyTHon"))
```

```python
def common(a, b):

    result = []

    for i in a:
        if i in b:
            result.append(i)

    return result

print(common([1, 2, 3], [2, 3, 4]))
```

```python
def digit_sum(num):

    total = 0

    for i in str(num):
        total += int(i)

    return total

print(digit_sum(1234))
```

```python
def armstrong(num):

    total = 0
    power = len(str(num))

    for i in str(num):
        total += int(i) ** power

    if total == num:
        return "Armstrong Number"

    return "Not Armstrong Number"

print(armstrong(153))
```
