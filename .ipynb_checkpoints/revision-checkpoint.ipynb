### How do I modify a string in python? 

```python
#concatination(+)
s='hello'
s1=s+'sir'
print(s1)
```

```python
#repetation(*)
s1=s*2
print(s1)
```

```python
# using slicing
str_='sun'
new_str='j'+str_[1:]
print(new_str)
```

```python
# type casting 
# In "" .join(lst), the empty string "" acts as a separator. 
# It joins all list elements together without adding spaces or symbols between them.
str_text='sun'
list_text=list(str_text)
# print(list_text)
list_text[0]='j'
new_str=''.join(list_text)
print(new_str)
```

### 19. Write a program to print all keywords in python? 
* In Python, a keyword is a reserved word that has a special meaning in the language and cannot be used as an identifier (like a variable name, function name, etc.).


```python
import keyword
x=keyword.kwlist
print('Total keywords=',len(x))
print('Keywords=',x)
```

### 20. Write a program to print punctuation in python.

### 21. What is a token in python? 
In Python, a token is the smallest unit of the source code that the Python
interpreter recognizes during the process of lexical analysis (the first step in code
compilation or interpretation). Each token represents a meaningful element in
Python, such as
1. Keywords.
2. Punctuation/delimiters.
3. Identifiers.
4. Operators.
5. Literals


### inbuilt fn vs inbuilt method in python?
Inbuilt functions are predefined functions provided by Python that can be used directly, while inbuilt methods are functions associated with specific objects or datatypes and are called using the dot operator.

### 24. What is the list in python? Explain list methods with examples. 
 We use a list in Python when we want to store multiple objects together in an ordered and mutable collection that also allows duplicate values.

```python
#Different memory address so, list is mutable object.
list_1=['suraj','sun']
list_2=['rohan','shohan']
print(id(list_1),id(list_2))
```

```python
# Duplicates are allow 
list_1=['ram',40,29,13,{'age':40},35] #homogeneous and heterogeneous both objects are allow
print(list_1)
```

```python
# Order preserved 
list_1=['ram',40,29,13,{'age':40},35]
x=0
for i in list_1: 
    print('list_1[{}]='.format(x),i)
    x=x+1
```

* In Python, .format() is used to insert values into a string dynamically.

```python
demo=[43,27,445,765,'age','name',78]
x=0
y='index'
for i in demo: 
    print('insert[{}]'.format(x,y),i)
    x=x+1
```
